## Easy Tasks:

### **Task 1**: Create a simple line plot using Matplotlib.
   - Use `x = [1, 2, 3, 4, 5]` and `y = [1, 4, 9, 16, 25]`.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")

plt.show()
```

### Task 2: Create a bar chart to display sales data of three products.
  - Use products = ['Television', 'Laptop', 'Mobile'] and sales = [150, 200, 300].

```python
products = ['Television', 'Laptop', 'Mobile']
sales = [150, 200, 300]

plt.bar(products, sales)

plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Sales Data of Three Products")

plt.show()
```

### Task 3: Create a scatter plot with random data for x and y.
  - Use x = [5, 7, 8, 7, 2] and y = [99, 86, 87, 88, 100].

```python
x = [5, 7, 8, 7, 2]
y = [99, 86, 87, 88, 100]

plt.scatter(x, y)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")

plt.show()
```

## Medium-Level Tasks:

### Task 4: Create a customized line plot with labels for axes and a title.
  - Use x = [1, 2, 3, 4] and y = [10, 20, 25, 30] and add a title and labels for x-axis and y-axis.

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)

plt.xlabel("(Time)")
plt.ylabel("(Value)")
plt.title("Customized Line Plot")

plt.show()
```

### Task 5: Create a bar chart with different colors for each bar.
  - Use categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 8].

```python
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]
colors = ['red', 'green', 'blue', 'orange']

plt.bar(categories, values, color=colors)

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart with Different Colors")

plt.show()
```

### Task 6: Create a subplot with two charts:
  - One line plot for x = [1, 2, 3, 4] and y = [10, 20, 25, 30].
  - One bar chart for categories = ['A', 'B', 'C'] and values = [50, 70, 30].

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

categories = ['A', 'B', 'C']
values = [50, 70, 30]

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.subplot(1, 2, 2)
plt.bar(categories, values, color=['red', 'green', 'blue'])
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.show()
```

Task 7: Create a pie chart to represent the percentage of students with grades A, B, C, and F.
  - Use labels = ['A', 'B', 'C', 'F'] and sizes = [15, 30, 45, 10].

```python
labels = ['A', 'B', 'C', 'F']
sizes = [15, 30, 45, 10]
explode = [0.1, 0, 0, 0]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode)

plt.title("Distribution of Students' Grades")

plt.show()
```

## Challenging Tasks:

### Task 8: Create a line plot and customize it with different colors, markers, and line styles.
  - Use x = [1, 2, 3, 4] and y = [10, 20, 25, 30] with a red line, circular markers, and a dashed line style.

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, color='red', marker='o', linestyle='--', linewidth=2, markersize=8)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")

plt.show()
```

### Task 9: Create a bar chart and customize the width and alignment of the bars.
  - Use categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 8].

```python
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.figure(figsize=(8, 5))

bar_width = 0.4
bar_positions = range(len(categories))
plt.bar(bar_positions, values, width=bar_width, align='center', color='skyblue')

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")

plt.xticks(bar_positions, categories)
plt.show()
```

### Task 10: Create a figure with two subplots:

  - One plot shows a scatter plot of x = [1, 2, 3, 4] and y = [10, 20, 25, 30].
  - Another plot shows a pie chart for labels = ['A', 'B', 'C'] and sizes = [40, 30, 30].

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

labels = ['A', 'B', 'C']
sizes = [40, 30, 30]
explode = [0.1, 0, 0]

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(x, y, color='blue', marker='o')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.subplot(1, 2, 2)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode, startangle=90)
plt.title("Pie Chart")

plt.show()
```
